```python
import pandas as pd

# 尝试使用其他编码格式打开文件
data1 = pd.read_csv('val_rus - 副本(1).csv')

# 导入第二个数据集
data2 = pd.read_csv('updated_gas_oil_tasks .csv')

# 输出前5行数据查看
print("Dataset 1:")
print(data1.head())

print("\nDataset 2:")
print(data2.head())
```

```python
import pandas as pd
import re

# 假设 data1 已经被读取

# 定义一个函数，用于检查一行是否包含俄语文本
def contains_russian(text):
    # 检查是否包含俄语字符（Unicode 范围）
    return bool(re.search('[а-яА-ЯёЁ]', str(text)))

# 使用 apply() 函数应用到 'standard_task_id' 列
data1_russian = data1[data1['standard_task_id'].apply(contains_russian)]

# 输出包含俄语文本的行
print(data1_russian.head())
```

```python
import pandas as pd

# 假设 data1 和 data2 已经被加载
# 使用 merge 进行连接，查找标准任务 ID 是否出现在 data2 的 general_name 列中
merged_data = pd.merge(data1, data2[['general_name']], left_on='standard_task_id', right_on='general_name', how='inner')

# 输出合并后的数据，查看匹配项
print(merged_data.head())
```
